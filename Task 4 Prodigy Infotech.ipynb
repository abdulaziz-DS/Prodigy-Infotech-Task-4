```python
import pandas as pd
df = pd.read_csv(r'C:\Users\Abdul\Downloads\twitter_training.csv')
df
```

```python
# Display the first few rows of the dataset
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Drop rows with missing values if necessary
df = df.dropna()
df
```

```python
print(df.columns)
```

```python
df.columns = ['ID', 'Topic', 'Sentiment', 'Tweet']
```

```python
# Check the distribution of sentiments
sentiment_counts = df['Sentiment'].value_counts()
print(sentiment_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the visualization
sns.set(style='whitegrid')

# Create a bar plot for sentiment distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='viridis')
plt.title('Sentiment Distribution in Twitter Data')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Assuming there is a 'topic' or 'brand' column
topic_sentiment = df.groupby('Topic')['Sentiment'].value_counts().unstack()

# Visualize sentiment patterns by topic
plt.figure(figsize=(12, 8))
topic_sentiment.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Sentiment Patterns by Topic')
plt.xlabel('Topic')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')
plt.show()
```
